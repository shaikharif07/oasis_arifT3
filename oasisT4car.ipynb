```python
# create a correlation matrix of the "train" DataFrame
hig_corr = car.corr()

# identify the features that have a correlation coefficient with "SalePrice"
# that is greater than or equal to 0.5
hig_corr_features = hig_corr.index[abs(hig_corr["price"]) >= 0.5]
hig_corr_features
```

```python
plt.figure(figsize=(12,10))
ax = sns.heatmap(car[hig_corr_features].corr(), cmap='coolwarm',annot=True,linewidth=3)
```

```python
car = car[hig_corr_features]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(car.iloc[:, :-1], 
                                                    car.iloc[:, -1], 
                                                    test_size = 0.3, 
                                                    random_state = 42)
```

```python
X_train.info()
```

```python
standardScaler = StandardScaler()
standardScaler.fit(X_train)
X_train = standardScaler.transform(X_train)
X_test = standardScaler.transform(X_test)
```

```python
linearRegression = LinearRegression()
linearRegression.fit(X_train, y_train)
y_pred = linearRegression.predict(X_test)
r2_score(y_test, y_pred)
```

```python
rf = RandomForestRegressor(n_estimators = 100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
r2_score(y_test, y_pred)
```
